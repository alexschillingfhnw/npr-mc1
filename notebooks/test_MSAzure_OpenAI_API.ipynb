```python
import os
import sys
from openai import AzureOpenAI

# Get the current working directory
cwd = os.getcwd()

# Construct the path to the src folder dynamically
credentials_folder = os.path.join(cwd, '../src/')

# Normalize the path to ensure it is absolute
credentials_folder = os.path.abspath(credentials_folder)

# Add the credentials folder to sys.path
sys.path.append(credentials_folder)

deployment_name='gpt-4' #This will correspond to the custom name you chose for your deployment when you deployed a model. 

print(credentials_folder)
```

```python
# Now you can import the credentials module
import credentials

# Use the values from the credentials
api_key = credentials.api_key  # Assuming you have api_key defined in credentials.py
azure_endpoint = credentials.azure_endpoint  # Assuming you have azure_endpoint defined
```

```python
client = AzureOpenAI(
    api_key=api_key,
    api_version="2023-12-01-preview",
    azure_endpoint=azure_endpoint
)
```

# ChatGPT

```python
start_phrase = 'What is your name? '

response = client.chat.completions.create(
    messages=[
            {"role": "system", "content": "Your name is Jürg and you are very, very polite assistant and you talk like a pirate."},
            {"role": "user", "content": start_phrase}
    ],
    model = deployment_name,
    #max_tokens = 20
)

response
print(f"Q: {start_phrase}\nA: {response.choices[0].message.content}")
```

# Embeddings

```python
def generate_embeddings(text, model="text-embedding-3-large"): # model = "deployment_name"
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
emb = generate_embeddings("What is my embedding?")
len(emb), emb
```

```python
emb2 = generate_embeddings("What is my embedding?", model="text-embedding-ada-002")
len(emb2), emb2
```
